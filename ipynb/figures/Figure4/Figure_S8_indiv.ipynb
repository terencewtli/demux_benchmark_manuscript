```r
library(ggplot2)
library(RColorBrewer)
library(cowplot)
library(gridExtra)
library(ggsci)
library(scales)
library(viridis)

library(ComplexHeatmap)
library(circlize)
library(tidyverse)
```

```r
projdir <- '/u/project/cluo/terencew/igvf/pilot/multiome'
```

```r
igvf_gex <- read.table(paste0(projdir, '/csv/final_figures/figure4/igvf_donor_gex.csv'), sep='\t', header=1, row.names=1)
igvf_atac <- read.table(paste0(projdir, '/csv/final_figures/figure4/igvf_donor_atac.csv'), sep='\t', header=1, row.names=1)
adelus_gex <- read.table(paste0(projdir, '/csv/final_figures/figure4/adelus_donor_gex.csv'), sep='\t', header=1, row.names=1)
adelus_atac <- read.table(paste0(projdir, '/csv/final_figures/figure4/adelus_donor_atac.csv'), sep='\t', header=1, row.names=1)
```

```r
head(igvf_gex)
```

```r
options(repr.plot.width=8, repr.plot.height=6)
cell_fun <- function(j, i, x, y, width, height, fill) {
        grid.text(sprintf("%.2f", igvf_gex[i, j]), x, y, gp = gpar(fontsize = 12,
                                                                      fontface = 'bold'))}
igvf_heatmap <- Heatmap(igvf_gex,
                        name='Singleton-donor\n correlation',
                    row_names_side = "left",
                    column_names_side = "bottom",
#                     top_annotation = gex_column_ha,
#                     col = colorRamp2(c(0.5, 0.75, 1), c("light blue", "white", "maroon")),
                    col = colorRamp2(c(0, 1), c("white", "red")),
                    column_title = 'Stem cell RNA',
                    row_title = '',
                    show_heatmap_legend = TRUE,
                    show_column_names=TRUE,
                    show_row_names=TRUE,
#                     row_labels = gex_order,
#                     row_split = gex_methods$geno_type,
                    row_title_gp = gpar(fontsize = 22),
                    column_title_gp = gpar(fontsize = 22),
#                     column_split = amb_rate,
                    cell_fun = cell_fun,
                    heatmap_width = unit(15, "cm"),
                    heatmap_height = unit(15, "cm"),
                    cluster_rows = FALSE,
                    cluster_columns = FALSE)
igvf_heatmap
```
